```python
import os
import pandas as pd
filenames = [x for x in os.listdir('./') if 'xls' in x and '$' not in x]
```

```python
len(filenames)
```

```python
df = pd.read_excel(filenames[0], sheet_name=0)
a = df.as_matrix()
```

```python
index = [i for i in range(len(a[:,0])) if a[i,0] =='مركز الإقتراع']
df2 = pd.DataFrame(a[index[0]:,:])
writer = pd.ExcelWriter('a.xlsx')
df2.to_excel('a.xlsx', sheet_name="Sheet1Name",startrow=0, startcol=0,header=False,index=False)
```

```python
for file in filenames:
    df = pd.read_excel(file, sheet_name=0)
    a = df.as_matrix()
    index = [i for i in range(len(a[:,0])) if a[i,0] =='مركز الإقتراع']
    df2 = pd.DataFrame(a[index[0]:,:])
    df2.to_excel('clean/'+file, sheet_name="Sheet1Name",startrow=0, startcol=0,header=False,index=False)
    print(index)
```
